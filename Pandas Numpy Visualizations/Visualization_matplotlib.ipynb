```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
# Line Plot
x = np.linspace(0,10,200)
```

```python
y = np.sin(x)
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,'--b')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("sin curve")
plt.show()
```

```python
# Scatter Plot

x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x,y)
```

```python
# Control scatter plot parametere

colors = np.random.rand(50)
sizes = 1000*np.random.rand(50)

plt.scatter(x,y,c = colors,s = sizes)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
# bar plot:

x = ['a','b','c','d']
y = np.linspace(0,1,4)

plt.bar(x,y)
```

```python
# horizontal bar plot
plt.barh(x,y)
```

```python
# histogram plot

data = [1,2,3,3,5,3,2,2,1,1,6,7,8,8,7,6,7,9,9]

plt.hist(data)
```

```python
# as randn gives std normal distribution, we can see that from the graph
 
x= np.random.randn(10000)
plt.hist(x)
```

```python
# 3D plot

x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(x,y,z)
plt.show()
```
