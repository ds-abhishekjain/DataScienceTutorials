```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
# Line Plot
x = np.linspace(0,10,200)
```

```python
y = np.sin(x)
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,'--b')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("sin curve")
plt.show()
```

```python
# Scatter Plot

x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x,y)
```

```python
# Control scatter plot parametere

colors = np.random.rand(50)
sizes = 1000*np.random.rand(50)

plt.scatter(x,y,c = colors,s = sizes)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
# bar plot:

x = ['a','b','c','d']
y = np.linspace(0,1,4)

plt.bar(x,y)
```

```python
# horizontal bar plot
plt.barh(x,y)
```
